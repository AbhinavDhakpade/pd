```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
df["Attrition"]=df["Attrition"].map({"Yes":1,"No":0})
```

```python
df
```

```python
df=df[['Age','DistanceFromHome','Attrition','Education','TotalWorkingYears']]
```

```python
df
```

```python
x=df[['Age','DistanceFromHome','Education','TotalWorkingYears']]
```

```python
y=df['Attrition']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print("Acc-scor:",accuracy_score(lr_pred,y_test))
print("conf-met:",confusion_matrix(lr_pred,y_test))
```
